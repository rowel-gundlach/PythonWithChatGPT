### Example use case
Retrieve vehicle information based on a license plate.

```python
import urllib.request
import json
import pandas as pd

kenteken = 'L173DH'
url = f"https://opendata.rdw.nl/resource/m9d7-ebf2.json?kenteken={kenteken}"

response = urllib.request.urlopen(url)
data = json.loads(response.read().decode())

#print(data)

df = pd.DataFrame(data)
print(df)
```

## 2. Connect to Open-Meteo API

The **Open-Meteo API** is a free weather API.

It allows you to retrieve:

- current temperature
- wind speed
- hourly forecasts
- precipitation
- historical weather

### Example use case
Retrieve the current weather for Amsterdam.

```python
import urllib.request
import json

# Amsterdam
url = "https://api.open-meteo.com/v1/forecast?latitude=52.37&longitude=4.90&current_weather=true"

response = urllib.request.urlopen(url) 
weather = json.loads(response.read().decode())
    
print(weather)

print("Temperature:", weather["current_weather"]["temperature"], "°C")
print("Wind speed:", weather["current_weather"]["windspeed"], "km/h")
```

Retrieve the weather of the last few days of Amsterdam, put it in a dataframe

```python
import urllib.request
import json
from datetime import datetime, timedelta
import pandas as pd

# Amsterdam coordinates
latitude = 52.37
longitude = 4.90

# Calculate dates for the last 5 days
end_date = datetime.today()
start_date = end_date - timedelta(days=5)

# Fetch historical weather data
url = f"https://archive-api.open-meteo.com/v1/archive?latitude={latitude}&longitude={longitude}&start_date={start_date.date()}&end_date={end_date.date()}&daily=temperature_2m_max,temperature_2m_min&timezone=Europe/Amsterdam"

response = urllib.request.urlopen(url)
weather = json.loads(response.read().decode())

# Create a Pandas DataFrame
df = pd.DataFrame({
    "date": weather["daily"]["time"],
    "temp_max": weather["daily"]["temperature_2m_max"],
    "temp_min": weather["daily"]["temperature_2m_min"]
})

print(df)
```

## 4. Connect to ECB API

The **ECB API** provides financial data from the European Central Bank.

A common use case is retrieving **currency exchange rates**.

Examples include:

- EUR to USD
- EUR to GBP
- historical exchange rates

### Example use case
Retrieve exchange rates from the ECB.

```python
import urllib.request
import urllib.parse
import pandas as pd
from io import StringIO

BASE = "https://data-api.ecb.europa.eu/service/data"

params = urllib.parse.urlencode({
    "startPeriod": "2026-03-25",
    "format": "csvdata",
})

url = f"{BASE}/EXR/D..EUR.SP00.A?{params}"

response = urllib.request.urlopen(url, timeout=15)
text = response.read().decode()

df = pd.read_csv(StringIO(text))
df.to_csv("ecb_exchange_rates.xlsx", index=False)
print(f"Saved {len(df)} rows to ecb_exchange_rates.xlsx")
```
