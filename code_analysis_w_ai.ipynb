### Exercise 2: What is this code checking for?

This code evaluates some kind of condition.

### Your task:
- Use AI to analyze the code
- Determine **what condition is being checked**

```python
def check_word_strength(word):
    score = 0

    if len(word) >= 8:
        score += 1
    if any(c.islower() for c in word):
        score += 1
    if any(c.isupper() for c in word):
        score += 1
    if any(c.isdigit() for c in word):
        score += 1
    if any(c in "!@#$%^&*" for c in word):
        score += 1

    return score


def get_feedback(score):
    if score <= 2:
        return "Weak"
    elif score == 3:
        return "Moderate"
    else:
        return "Strong"


# Input + execution (for notebook use)
word = input("Enter a word: ")

score = check_word_strength(word)
feedback = get_feedback(score)

print(f"\nScore: {score}/5")
print(f"Result: {feedback}")
```

### Exercise 3: Player Rankings System

This code keeps track of rankings of players after games.

### Your task:
- Analyze how the ranking system works
- Identify what kind of system this might represent

```python
# Initial players and ratings
players = {
    "Player1": 1000,
    "Player2": 1000,
    "Player3": 1000,
    "Player4": 1000
}

K = 32


def expected_score(rating_a, rating_b):
    return 1 / (1 + 10 ** ((rating_b - rating_a) / 400))


def update_elo(player_a, player_b):
    rating_a = players[player_a]
    rating_b = players[player_b]

    expected_a = expected_score(rating_a, rating_b)
    expected_b = expected_score(rating_b, rating_a)

    players[player_a] = round(rating_a + K * (1 - expected_a))
    players[player_b] = round(rating_b + K * (0 - expected_b))


def print_ranking():
    print("\nCurrent Rankings:")

    # Convert dictionary to list of (name, rating)
    ranking = list(players.items())

    # Print results
    for item in ranking:
        name = item[0]
        rating = item[1]
        print(name + ": " + str(rating))


# Show initial rankings

player_a = input("Winner: ")

player_b = input("Loser: ")

if player_a not in players or player_b not in players:
    print("Invalid player name.")
else:
    update_elo(player_a, player_b)
    
print_ranking()
```

### Exercise 4: Distance Between Places

This code calculates the distance between locations on Earth.

However… the results seem slightly off 👀

### Your task:
- Use AI to analyze the formula and logic
- Figure out **why the result may not be accurate**

```python
import math

def haversine(lat1, lon1, lat2, lon2):
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = math.cos(dlat)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon)**2
    c = 2 * math.asin(math.sqrt(a))

    R = 6371  # Earth radius in km
    return R * c


# Paris and Amsterdam coordinates
lat_paris, lon_paris = 48.8566, 2.3522
lat_amsterdam, lon_amsterdam = 52.3676, 4.9041

distance = haversine(lat_paris, lon_paris, lat_amsterdam, lon_amsterdam)
print(f"Distance: {distance:.2f} km")
```
