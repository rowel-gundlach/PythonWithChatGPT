## Below examples of code with typical mistakes. We use AI to spot and fix these mistakes

```python
# Goal: print the names of students who passed (grade >= 6)

students = [
    ("Alice", "8"),
    ("Bob", "5"),
    ("Clara", "9"),
    ("David", "7"),
    ("Eva", "4"),
]

passed = []

for name, grade in students:
    if grade >= 6:
        passed.append(name)

print("Students who passed:")
for name in passed:
    print("-", name)
```

```python
# Goal: You input a word (in this case "crane")
# You get 6 turns to guess a word
# If the letter is correct and in the correct place it is printed
# If the letter is at the wrong place but is in the word, it gets a "?" next to it.


SECRET = "crane"
MAX_TURNS = 6

print("Guess the 5-letter word! You have 6 attempts.\n")

guess = input("Attempt 1: ").lower().strip()

for attempt in range(MAX_TURNS):        
    result = ""
    for i in range(5):
        if guess[i] == SECRET[i]:
            result += "[ " + guess[i].upper() + " ] "
        elif guess[i] in SECRET:
            result += " " + guess[i].upper() + "? "
        else:
            result += " _  "

    print(f"Attempt {attempt + 1}: " + result.strip() + "\n")

    if guess == SECRET:
        print(f"Correct! You got it in {attempt + 1} attempt(s)!")
        break
else:
    print(f"Out of attempts! The word was: {SECRET.upper()}")
```

```python
# Goal: Extract mail adresses
# Input an email field and extract only th email adress

import re

def clean_email(email):

    email = email.strip().lower()
    match = re.search(r'@[\w.]+', email)  
    return match.group(0) if match else None

clean_email('mail:akad@hotmail.com')
```

```python
# Goal: print FizzBuzz for numbers 1 to 20
# - Print 'FizzBuzz' if divisible by both 3 and 5
# - Print 'Fizz' if divisible by 3
# - Print 'Buzz' if divisible by 5
# - Otherwise print the number

for n in range(1, 21):
    if n % 3 == 0 or n % 5 == 0:   
        print("FizzBuzz")
    elif n % 3 == 0:
        print("Fizz")
    elif n % 5 == 0:
        print("Buzz")
    else:
        print(n)
```

```python
# Goal: convert a temperature from Celsius to Fahrenheit

def celsius_to_fahrenheit(c):
    result = (c * 9/5) + 32


temp_c = 22
temp_f = celsius_to_fahrenheit(temp_c)

print(f"{temp_c}°C is {temp_f}°F")
print(f"Is it warm? {temp_f > 68}")
```
